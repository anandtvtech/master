Problem
IT Industry records the amount of time a software engineer needs to fix a bug in the initial phase of software development in ‘debugging.csv’

Let
X = time needed to fix bugs

X is a continuous random variable. Les see the distribution of X and answer below questions.

1. Find the probability that a randomly selected software debugging requires less than 3 hours
2. Find the probability that a randomly selected software debugging requires more than 2 hours
3. Find the 50th percentile of software debugging time

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
import scipy.stats as stats

debugging = pd.read_csv('/content/drive/MyDrive/ColabNotebooks/Probability distribution/debugging.csv')
debugging.head()
```

```python
plt.hist(debugging['Time Taken to fix the bug'], density=True)
plt.axhline(1/4, color='red')
plt.xlabel('Time rquired for bug fixing')
plt.ylabel('Probablity')
plt.title('Data distribution')
plt.show()
```

Use dist to check the density


Insight:(Below)
As you can see from the above plot that all the values between 1 and  5 are having almost equal probability, we are going to use continuous uniform distribution. We need to decide the endpoint.
Here endpoints are 1 and 5

X~U(1,5)

```python
sns.distplot(debugging['Time Taken to fix the bug'])
plt.show()
```

```python
from scipy.stats import uniform
x = np.linspace(1, 5, 50)
probs = uniform.pdf(x, loc=1, scale=4) #from 1 to 5 elements so loc 1 and scale 4 
probs
```

Find prob of max time of 3 hours
So Find CDF
Fx(X) = P(X<=x)
P(X<=3)

```python
x1 = np.linspace(1,3,25)

plt.plot(x, probs)
plt.fill_between(x, probs)
plt.fill_between(x1, uniform.pdf(x=x1,loc=1,scale=4),color="r")
plt.xlabel("Time required for bug fixing")
plt.ylabel("Probablity")
plt.title("Continuous Uniform Distribution: X~U(1,5)")
plt.show()

uniform.cdf(x=3, loc=1, scale=4) #P(X<=3)
```

2)More than 2 hours P(x>2)
So Find cdf for 1-P(X=2)

```python
x1 = np.linspace(2,5,25) #To find area between 2 and above and max range is 5

plt.plot(x, probs)
plt.fill_between(x, probs)
plt.fill_between(x1, uniform.pdf(x=x1,loc=1,scale=4),color="r")
plt.xlabel("Time required for bug fixing")
plt.ylabel("Probablity")
plt.title("Continuous Uniform Distribution: X~U(1,5)")
plt.show()

1 - uniform.cdf(x=2, loc=1, scale=4) #P(X>2)
```

3)50% percentile of debugging time
Note:ppf() is used ot calculate percentile point given probablity. Its opposite to cdf.

```python
uniform.ppf(q=0.5,loc=1,scale=4) #to find 50 % between min 1 and max 5 value
```

Conclusion:
a)There is a 50% chance that a randomly selected software debugging requires less than three hours

b)There is an 75% chance that a randomly selected software debugging requires more than two hours

c)The 50th percentile of software debugging time is 3 hours
