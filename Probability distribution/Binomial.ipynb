80% of all the visitors to Lavista Museum end up buying souvenirs from the souvenir shop at the museum. On the coming Sunday, If a random sample of 10 visitors is picked

1. Find the probability that every visitor will end up buying from the souvenir shop
2. Find the probability that a maximum of 7 visitors will buy souvenirs from the souvenir shop

Solution:
Lets check first whether we satisfy the assumption of binomial distribution

1. Visitor buy from shop or not (success or failure)
2. No of trials is fixed - 10 visitors
3. Each trial is independent
4. Prob of success is same for each trial . Pro is 0.8%

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import scipy.stats as stats
n = 10
p = 0.8
k = np.arange(0,11)
k
````

````python
from scipy.stats import binom

binomial  = binom.pmf(k=k, n=n,p=p)
binomial
````

````python
plt.bar(k, binomial)
plt.title("Binomial n=%i,p=%.2f" %(n,p), fontsize=15)
plt.xlabel("Number of success")
plt.ylabel("Probablity of Success")
plt.show()
````

1) For 10 person

````python
binomial[10]
````

For Max 7 persons

````python
bar1 = plt.bar(k, binomial)
plt.title("Binomial n=%i,p=%.2f" %(n,p), fontsize=15)
plt.xlabel("Number of success")
plt.ylabel("Probablity of Success")
for i in range(0,8):
  bar1[i].set_color('r')
plt.show()
````

````python
binom.cdf(k=7,n=n,p=p)
````

3)Prob of atleast 3

ie. 1-(p<=3)

````python
1-binom.cdf(k=3,n=n,p=p)
````

Conclusion:
There is 10.74 % chance that every visitor will end up buying souvenirs from the souvenir shop.

There is 32.22% chance that a maximum of 7 visitors out of 10 will buy
